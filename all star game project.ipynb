```python
nba_df_HLP = nba_df.loc[(nba_df['PTS']>=400)& (nba_df['AST']>=50)& (nba_df['REB']>=50)& (nba_df['GP']>=15)]
nba_df_HLP
```

```python
western_conferences=['LAL','GSW','MIN','UTA','NOP','SAC','POR','PHX','DAL','OKC','SAS','MEM','HOU','LAC','DEN']
# nba_df_HLP.loc[(nba_df_HLP['TEAM_ABBREVIATION']== '' )]
nba_df_HLPW=nba_df_HLP.loc[(nba_df_HLP['TEAM_ABBREVIATION'].isin(western_conferences))]
nba_df_HLPW
```

```python
#E=sum of points, assits, rebounds, steals and blocks of each player.
nba_df_HLPW['E']= nba_df_HLPW['REB']+nba_df_HLPW['AST']+nba_df_HLPW['PTS']+nba_df_HLPW['STL']+nba_df_HLPW['BLK']
#Impact Per Game
nba_df_HLPW['IPG']=nba_df_HLPW['E']/nba_df_HLPW['GP']
nba_df_HLPW
```

```python
nba_df_HLPW_selected = nba_df_HLPW.loc[nba_df_HLPW['IPG']>=36]
nba_df_HLPW_selected
```

```python
estern_conferences=['CHI','CHA','NYK','MIA','WAS','ATL','MIL','CLE','PHI','TOR','BOS','BKN','ORL','IND','DET']
# nba_df_HLP.loc[(nba_df_HLP['TEAM_ABBREVIATION']== '' )]
nba_df_HLPE=nba_df_HLP.loc[(nba_df_HLP['TEAM_ABBREVIATION'].isin(estern_conferences))]
nba_df_HLPE
```

```python
#E=sum of points, assits, rebounds, steals and blocks of each player.
nba_df_HLPE['E']= nba_df_HLPE['REB']+nba_df_HLPE['AST']+nba_df_HLPE['PTS']+nba_df_HLPE['STL']+nba_df_HLPE['BLK']
#Impact Per Game
nba_df_HLPE['IPG']=nba_df_HLPE['E']/nba_df_HLPE['GP']
nba_df_HLPE
```

```python
nba_df_HLPE_selected = nba_df_HLPE.loc[nba_df_HLPE['IPG']>=38.5]
nba_df_HLPE_selected
```
